```python
#Ввести
print('🎮 ДОБРО ПОЖАЛОВАТЬ В СОЗДАТЕЛЬ ПЕРСОНАЖА!')
print('=' * 60)
n = input('Введите имя: ').strip()
a = input('Введите возраст: ').strip()
c = input('Введите город: ').strip()
py = input('Введите язык программирования: ').strip()
skills_list = ['Креативность','Аналитика', 'Программирование','Логика', 'SQL', 'Data Scientist']
person = {'name': n,
      'age': a,
      'city': c ,
      'language': py,
      'skills': skills_list
      }
quote = 'Андерсен, не говори вслух, ты понижаешь IQ всей улицы!'

#Обработка данных
a = int(a)
n = n.capitalize()
c = c.capitalize()
py = py.upper()
count = len(skills_list)

print('=' * 60)
print('\n🎮 ВЫБОР СЛОЖНОСТИ:')
print('=' * 60)
diff = input('Выберите уровень сложности игры: Лёгкий/Средний/Сложный').strip().lower()
exp = 0
if diff == 'лёгкий':
  exp += 50
  print(f'✅ Лёгкий уровень! +{exp} XP')
elif diff == 'средний':
  exp += 100
  print(f'✅ Средний уровень! +{exp} XP')
elif diff == 'сложный':
  exp += 200
  print(f'✅ Сложный уровень! +{exp} XP')


if py == 'PYTHON':
  print('+10 баллов к интелекту')
  print('🐍 Отличный выбор! Python — язык магии!')

if 'Командная работа' in skills_list:
  print('Правильный выбор')
else:
  skills_list.append('Командная работа')
  print('💡 Совет: добавьте навык Командная работа — он важен для разработчика!')

print('------------- ВЫБОР СПЕЦИАЛИЗАЦИИ:-------------------')
spe = input('Выберите специализацию (web/mobile/data science/gamedev/универсал):')
if spe == 'web':
    specialization_name = "Веб-разработчик"
    print(f'Специализация: {specialization_name}')
elif spe == 'mobile':
    specialization_name = "Мобильный разработчик"
    print(f'Специализация: {specialization_name}')
elif spe == 'data science' or spe == 'data_science' or spe == 'datascience':
    specialization_name = "Data Scientist"
    print(f'Специализация: {specialization_name}')
elif spe == 'gamedev':
    specialization_name = "Game Developer"
    print(f' Специализация: {specialization_name}')
else:
    specialization_name = "Универсал"
    print(f' Специализация: {specialization_name}')



#вывод
print('=' * 60)
print('ПРОФИЛЬ ПЕРСОНАЖА')
print('=' * 60)
print('Имя:', n)
print('Возраст:', a)
if 6 <= a < 18:
    print('Статус персонажа:⭐ Юный ученик')
elif 18 <= a <= 25:
    print('Статус персонажа:⭐⭐Студент')
elif 26 <= a <= 40:
    print('Статус персонажа:⭐⭐⭐⭐ Опытный разработчик')
else:
    print('Статус персонажа:⭐⭐⭐⭐⭐Мудрый наставник')
print('Город:', c)
print('Язык программирования:', py)
print(f'Специализация: {specialization_name}')
print(f'Очки опыта: {exp} XP')
print(f'Навыки({count}):')
for i in skills_list:
    print('•', i)
print(f'Любимая цитата: {quote}')
print('=' * 60)

```
